```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv('/content/Programming _ A Coding Fear Among the Students.csv')
```

```python
dataset.isnull().sum()
```

```python
dataset['Gender'].value_counts().plot(kind='barh')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
one=le.fit_transform(dataset['Gender'])
two=le.fit_transform(dataset['At which age you started learning programming?'])
three=le.fit_transform(dataset['How often do you code?'])
four=le.fit_transform(dataset['Does programming seem difficult or hard to you?'])
five=le.fit_transform(dataset['While coding, which one of these following method you take help from?'])
six=le.fit_transform(dataset['When you find any error in your code, what do you do?'])
seven=le.fit_transform(dataset['Do you think, "Coding is too difficult for me to learn?"'])
eight=le.fit_transform(dataset['What\'s the reason of getting anxious?'])
nine=le.fit_transform(dataset['What are the reasons behind this coding fears of yours'])
ten=le.fit_transform(dataset["Do you feel like you're trying your hard but not getting anywhere?"])
eleven=le.fit_transform(dataset['How do you feel when you start a code or any code related work?'])
twelve=le.fit_transform(dataset['When trying to code, which challenges you face the most?'])
thirteen=le.fit_transform(dataset['How do you feel seeking help from others'])
fourteen=le.fit_transform(dataset['How do you feel about coding assessments or exams?'])
fifteen=le.fit_transform(dataset['When working on a coding assignment or project, what do you struggle with the most?'])
sixteen=le.fit_transform(dataset['What do you think makes coding difficult to learn?'])
seventeen=le.fit_transform(dataset['How would you feel pairing with someone more experienced when learning coding?'])
eighteen=le.fit_transform(dataset['What resources do you think would help with learning coding?'])
nineteen=le.fit_transform(dataset['Do you think you\'re falling behind because of not knowing coding very well?'])
```

```python
dataset['Gender']=one
dataset['At which age you started learning programming?']=two
dataset['How often do you code?']=three
dataset['Does programming seem difficult or hard to you?']=four
dataset['While coding, which one of these following method you take help from?']=five
dataset['When you find any error in your code, what do you do?']=six
dataset['Do you think, "Coding is too difficult for me to learn?"']=seven
dataset['What\'s the reason of getting anxious?']=eight
dataset['What are the reasons behind this coding fears of yours']=nine
dataset["Do you feel like you're trying your hard but not getting anywhere?"]=ten
dataset['How do you feel when you start a code or any code related work?']=eleven
dataset['When trying to code, which challenges you face the most?']=twelve
dataset['How do you feel seeking help from others']=thirteen
dataset['How do you feel about coding assessments or exams?']=fourteen
dataset['When working on a coding assignment or project, what do you struggle with the most?']=fifteen
dataset['What do you think makes coding difficult to learn?']=sixteen
dataset['How would you feel pairing with someone more experienced when learning coding?']=seventeen
dataset['What resources do you think would help with learning coding?']=eighteen
dataset["Do you think you're falling behind because of not knowing coding very well?"]=nineteen
```

```python
dataset
```

```python
features=[
    'Gender',
    'At which age you started learning programming?',
    'How often do you code?',
    'Does programming seem difficult or hard to you?',
    'While coding, which one of these following method you take help from?',
    'When you find any error in your code, what do you do?',
    'Do you think, "Coding is too difficult for me to learn?"',
    'What\'s the reason of getting anxious?',
    'What are the reasons behind this coding fears of yours',
    'How do you feel when you start a code or any code related work?',
    'When trying to code, which challenges you face the most?',
    'How do you feel seeking help from others',
    'How do you feel about coding assessments or exams?',
    'When working on a coding assignment or project, what do you struggle with the most?',
    'What do you think makes coding difficult to learn?',
    'How would you feel pairing with someone more experienced when learning coding?',
    'What resources do you think would help with learning coding?',
    'Do you think you\'re falling behind because of not knowing coding very well?',
    'Do you feel like you\'re trying your hard but not getting anywhere?'
]
```

```python
y=dataset['Do you get anxious while coding? ']
```

```python
x=dataset[features]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
skb=SelectKBest(score_func=chi2)
```

```python
sb=skb.fit(x,y)
```

```python
sc=pd.DataFrame(sb.scores_)
```

```python
sc_c=pd.DataFrame(x.columns)
```

```python
pd.concat([sc_c,sc],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(xtrain,ytrain)
```

```python
dt.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,dt.predict(xtest))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytest,dt.predict(xtest))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,dt.predict(xtest)))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(xtrain,ytrain)
```

```python
rfc.predict(xtest)
```

```python
accuracy_score(ytest,rfc.predict(xtest))
```
